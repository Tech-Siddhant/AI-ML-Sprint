the only way to run your log aggregator is for you, the engineer, to manually open a terminal and type docker run. In the real world, an automated system needs to run this pipeline whenever a new log file arrives, or a web dashboard needs to trigger it to fetch live analytics
To do this, we are entering Week 3: Microservice Architecture. We are going to build a RESTful API using FastAPI

## What is API 
1. Application (The Machine)
This is the software you built. Your Week 1 Pandas Log Aggregator is the "Application." It does the heavy lifting of cleaning data.

2. Programming (The Audience)
Your application is currently designed for a human. You have to open a terminal, type docker run, and look at the screen. An API changes the audience. It allows other code (like a React website, a mobile app, or another server) to trigger your application automatically, without human intervention.

3. Interface (The Doorway)
An interface is simply a boundary where two different systems meet and communicate. A steering wheel is a human's interface to a car. A graphical website is a human's interface to a database. An API is a machine's interface to your code.
