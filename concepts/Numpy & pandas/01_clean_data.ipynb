## Step 1: Initial DataFrame
Notice that some rows have NaN (Not a Number) or None, which pandas automatically recognizes as missing values. However, row 5 contains the strings 'NA' and 'Missing', which are custom missing values that pandas currently just sees as normal text

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("messy_people_data.csv")
print (df)
```

#### 1. Drop rows with ANY standard missing values (np.nan or None)

```python
print (df.dropna())
```

#### 2. Drop rows only if ALL values in that row are missing

```python
print (df.dropna(how = "all"))
```

#### 3. Drop rows only if a specific column (like 'email') is missing a value

```python
print (df.dropna(subset = ['email']))
```

#### 4. Drop columns instead of rows (will return empty if all columns have at least one NaN)

```python
print (df.dropna(axis = 'columns'))
```

## Step 2: Checking for and Replacing Custom Missing Values
Pandas does not automatically know that the strings 'NA' and 'Missing' are supposed to be blank data
You can test checking for these and replacing them


#### View a mask (True/False) of what pandas currently considers a missing value

```python
print (df.isna())
```

####  Replace custom strings with proper numpy NaN values, permanently (inplace=True)

```python
df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
```

```python
print (df)
```

##### If you run dropna() now, it will successfully drop the row that had the custom strings print(df.dropna(how='all'))


## Step 3: Testing fillna()
If you are working with numerical data (like grades) and want missing values to default to a 0 instead of remaining blank, you use fillna()

#### Fill all NaN values with a 0

```python
print(df.fillna(0))
```

## Step 4: Testing dtypes and astype()
In our test data, the ages are strings (text), not numbers
. If you try to run df['age'].mean() right now, it will throw an error
. We need to cast it to a float

#### Check current data types (the columns will show up as 'object' / strings)

```python
print (df.dtypes)
```

#### Cast the 'age' column to a float (must be a float to handle np.nan values!)
as NAN is consider as a float so thats why we are not going to convert it into integer

```python
df['age'] = df['age'].astype(float)
```

```python
# verify that datatype is changed 
print(df.dtypes)
```

#### Now calculating mean age successfully 

```python
print("Average Age:", df['age'].mean())
```
