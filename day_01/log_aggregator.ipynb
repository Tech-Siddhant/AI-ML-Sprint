## Data Ingestion

```python
import pandas as pd 
df = pd.read_csv("massive_raw_logs.csv")
print ("-------raw data-------")
print (df)
```

## Cleaning Critical Data (The Drops)

```python
# A log without a timestamp is useless. Drop any row missing a timestamp.
df_cleaned = df.dropna(subset = ["timestamp"])
print ("------after cleaning missed timestamp------")
print (df_cleaned)
```

## Patching Non-Critical Data (The Fills)

```python
# We don't want to throw away a log just because the IP is missing. We patch it.
df_cleaned['ip_address'] = df_cleaned['ip_address'].fillna('UNKNOWN_IP')
df_cleaned['status_code'] = df_cleaned['status_code'].fillna(0) # 0 means unknown status
df_cleaned['status_code'] = df_cleaned['status_code'].fillna('None')

print("\n--- FULLY CLEANED DATA ---")
print(df_cleaned)
```
